```python
import yfinance as yf
import pandas as pd
from datetime import datetime
from google.colab import files

# Step 1: Define the stock ticker and date range
stock_ticker = "AAPL"  # Example: Apple Inc.
start_date = "2023-01-01"  # Start date for historical data
end_date = datetime.today().strftime('%Y-%m-%d')  # End date as today's date

# Step 2: Fetch historical stock data
print(f"Fetching data for {stock_ticker} from {start_date} to {end_date}...")
stock_data = yf.download(stock_ticker, start=start_date, end=end_date)

# Step 3: Display the first few rows of the data
if not stock_data.empty:
    print("Data fetched successfully! Here's a preview:")
    print(stock_data.head())
else:
    print("No data found. Please check the stock ticker or date range.")

# Step 4: Save the data to a CSV file
csv_file_name = f"{stock_ticker}_stock_data.csv"
stock_data.to_csv(csv_file_name)
print(f"Data saved to {csv_file_name}")

# Step 5: Download the CSV file to your local machine
files.download(csv_file_name)
```

```python
# Fetch real-time data for the current day
real_time_data = yf.Ticker(stock_ticker).history(period="1d")
print("Real-time stock data:")
print(real_time_data)
```
